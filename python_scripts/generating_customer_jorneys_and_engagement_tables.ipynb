```python
import pandas as pd
import numpy as np

customers_path = "/content/synthetic_bank_customers.csv"
products_path = "/content/banking_products.csv"
campaigns_path = "/content/banking_campaigns.csv"

customers_df = pd.read_csv(customers_path)
products_df = pd.read_csv(products_path)
campaigns_df = pd.read_csv(campaigns_path)

campaigns_df['start_date'] = pd.to_datetime(campaigns_df['start_date'])
campaigns_df['end_date'] = pd.to_datetime(campaigns_df['end_date'])

print("Number of Customers: ", len(customers_df))
display("Products: ", products_df)
display("Campaigns: ", campaigns_df)
```

```python
import pandas as pd
import numpy as np

# -----------------------------
# Realistic Customer Journey Table
# -----------------------------
num_journey_customers = 30000  # number of unique customer journeys to simulate
stages = ["Home Page", "Product Page", "Application Form", "Checkout", "Confirmation"]

# Function to generate a realistic journey for a single customer
def generate_journey(customer_id, product_id):
    journey = []
    # Random visit date over a 3-year period
    visit_date = pd.to_datetime('2023-01-01') + pd.to_timedelta(np.random.randint(0, 1095), unit='d')

    # Stage 1: Home Page - always viewed
    journey.append({
        "customer_id": customer_id,
        "product_id": product_id,
        "visit_date": visit_date,
        "stage": "Home Page",
        "action": "View",
        "duration_seconds": np.random.randint(10, 120)
    })

    # Stage 2: Product Page
    if np.random.rand() < 0.6:  # 60% of users click to product page
        journey.append({
            "customer_id": customer_id,
            "product_id": product_id,
            "visit_date": visit_date,
            "stage": "Product Page",
            "action": "Click",
            "duration_seconds": np.random.randint(15, 180)
        })

        # Stage 3+: Application Form / Checkout / Confirmation
        rand = np.random.rand()
        if rand < 0.3:
            journey.append({
                "customer_id": customer_id,
                "product_id": product_id,
                "visit_date": visit_date,
                "stage": "Application Form",
                "action": "Sign Up",
                "duration_seconds": np.random.randint(30, 300)
            })
        elif rand < 0.6:
            journey.append({
                "customer_id": customer_id,
                "product_id": product_id,
                "visit_date": visit_date,
                "stage": "Application Form",
                "action": "Drop Off",
                "duration_seconds": np.random.randint(5, 100)
            })
        else:
            journey.append({
                "customer_id": customer_id,
                "product_id": product_id,
                "visit_date": visit_date,
                "stage": "Checkout",
                "action": "Purchase",
                "duration_seconds": np.random.randint(60, 600)
            })

    return journey

# -----------------------------
# Generate journeys for multiple customers
# -----------------------------
# Randomly assign products to sampled customers
sample_customer_ids = np.random.choice(customers_df['customer_id'], num_journey_customers)
sample_product_ids = np.random.choice(products_df['product_id'], num_journey_customers)

all_journeys = []
for cust, prod in zip(sample_customer_ids, sample_product_ids):
    all_journeys.extend(generate_journey(cust, prod))

# Create DataFrame
customer_journey_df = pd.DataFrame(all_journeys)

# Add journey_id as the first column
customer_journey_df.insert(0, "journey_id", np.arange(1, len(customer_journey_df)+1))

display(customer_journey_df.head(20))
```

```python
file_name = "customer_journey_data.csv"
customer_journey_df.to_csv(file_name, index=False)
```

```python
# -----------------------------
# Engagement Table
# -----------------------------
num_engagements = 100000
content_types = ["Blog", "Facebook", "Instagram", "TikTok", "YouTube", "Newsletter"]
content_ids = np.arange(1, 501)  # 500 pieces of content

# Random content and type
engagement_content_ids = np.random.choice(content_ids, num_engagements)
engagement_content_types = np.random.choice(content_types, num_engagements)

# Engagement dates over 3 years
engagement_dates = pd.to_datetime('2023-01-01') + pd.to_timedelta(
    np.random.randint(0, 1095, num_engagements), unit='d'
)

likes = np.random.randint(100, 2000, num_engagements)
views = np.random.randint(1000, 5000, num_engagements)
clicks = np.minimum(np.random.randint(100, 1000, num_engagements), views)

# Assign campaigns based on engagement date
def assign_campaign_by_date(date, campaigns_df):
    active_campaigns = campaigns_df[
        (campaigns_df['start_date'] <= date) &
        (campaigns_df['end_date'] >= date)
    ]
    if len(active_campaigns) == 0:
        return np.nan
    else:
        return np.random.choice(active_campaigns['campaign_id'])

campaign_ids = [assign_campaign_by_date(d, campaigns_df) for d in engagement_dates]

# Assign random products (requires products_df already loaded)
product_ids = np.random.choice(products_df['product_id'], num_engagements)

# Create the engagement DataFrame
engagement_df = pd.DataFrame({
    "engagement_id": np.arange(1, num_engagements + 1),
    "content_id": engagement_content_ids,
    "content_type": engagement_content_types,
    "likes": likes,
    "views": views,
    "clicks": clicks,
    "engagement_date": engagement_dates,
    "campaign_id": campaign_ids,
    "product_id": product_ids,
    "views_clicks_combined": views.astype(str) + "-" + clicks.astype(str)
})

engagement_df['campaign_id'] = engagement_df['campaign_id'].fillna(-1)

display(engagement_df.head())
```

```python
filename = "banking_engagement_data.csv"

engagement_df.to_csv(filename, index=False)
```
